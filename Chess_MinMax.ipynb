```python
#getting white king indexes
def Get_WKing_Indexes(Board):
    for i in range(8):
        for j in range(8):
            if Board[i][j] != "  ":
                if Board[i][j].Color == "White" and Board[i][j].Name =="King" :
                    return (i,j)
     
    return (None, None)
                
            
#getting black king indexes
def Get_BKing_Indexes(Board):
    for i in range(8):
        for j in range(8):
            if Board[i][j] != "  ":
                if Board[i][j].Color == "Black" and Board[i][j].Name =="King" :
                    return (i,j)
    return (None,None)

```

```python
#class of rook piece
class Rook(Piece):
    def __init__(self, name,row, col, importance, color, gui, moved):
        super().__init__(name,row, col, importance, color, gui, moved)

    # Validating rook moves
    def Valid_Move(self,Row2,Col2,Board):
        #if the moves are not on the same row
        if Row2 != self.Row and Col2 != self.Col:
            return False
        #else if the same row
        if Row2 == self.Row:
            #if want to move right
            if Col2 > self.Col:
                for i in range(self.Col+1,Col2):
                    if (Board[Row2][i] != "  "):  # if no empty space
                        return False
            #if the move is left
            else:
                for i in range(Col2+1,self.Col):
                    if (Board[Row2][i] != "  "): # if no empty space
                        return False
        #else if same column
        if (Col2 == self.Col):
            #if want to move up
            if (Row2 > self.Row):
                for i in range(self.Row+1,Row2):
                    if (Board[i][Col2] != "  "): # if no empty space
                        return False
            #if want to move down
            else:
                for i in range(Row2+1,self.Row):
                    if (Board[i][Col2] != "  "): # if no empty space
                        return False
        #accessing the destination object
        obj = Board[Row2][Col2]
        # if blank
        if obj == "  ":
            return True
        
        # if the color is same, then return false
        elif obj.Color == self.Color:
            return False
        return True

 
    #function to Move rook
    def Move(self,Row2,Col2,Board):
        #creating temporary Row and Column for current indexes
        TempRow = self.Row
        TempCol = self.Col

        #changing the position of the piece
        self.Row = Row2
        self.Col = Col2
        
        self.Moved += 1

        #changing the position of the piece in the board
        Board[Row2][Col2] = self
        Board[TempRow][TempCol] = "  "
        return Board
```

```python
#class of knight piece
class Knight(Piece):
    def __init__(self, name,row, col, importance, color, gui, moved):
        super().__init__(name,row, col, importance, color, gui, moved)

    # Validating knight moves
    def Valid_Move(self,Row2,Col2,Board):
        #if move is up, left or right, do nothing
        if (abs(self.Row - Row2) == 2 and abs(self.Col - Col2) == 1):
            pass
        #if the move is down, left or right, do nothing
        elif (abs(self.Row - Row2) == 1 and abs(self.Col - Col2) == 2):
            pass
        else:
            # else return false
            return False
        
        #accessing the destination object
        obj = Board[Row2][Col2]

        if obj == "  ": # if empty 
            return True
        
        # if the color is same, then return false
        elif obj.Color == self.Color:
            return False
        return True

    # Move the Knight object in that index and put "  " in previous index
    def Move(self,Row2,Col2,Board):
        # store the previus position in temp variables
        tempRow = self.Row
        tempCol = self.Col

        self.Moved += 1
        #changing the position of the piece
        self.Row = Row2
        self.Col = Col2
        
        
        # Move the Knight object in that index and put "  " in previous index
        Board[Row2][Col2] = self
        Board[tempRow][tempCol] = "  "
        return Board
```

```python
#class of bishop piece
class Bishop(Piece):
    def __init__(self, name,row, col, importance, color, gui, moved):
        super().__init__(name,row, col, importance, color, gui, moved)

    # Validating bishop moves
    def Valid_Move(self,Row2,Col2,Board):
        # if moves not diagnol, return false
        if (abs(Row2-self.Row) != abs(Col2-self.Col)):
            return False
        #if the moves diagonal
        else:
            #if the move is up right
            if (Row2 > self.Row and Col2 > self.Col):
                for i in range(1,abs(Row2-self.Row)):
                    if (Board[self.Row+i][self.Col+i] != "  "): # if no empty space
                        return False        
            #if the move is up left
            elif (Row2 > self.Row and Col2 < self.Col):
                for i in range(1,abs(Row2-self.Row)):
                    if (Board[self.Row+i][self.Col-i] != "  "):
                        return False
            #if the move is down right
            elif (Row2 < self.Row and Col2 > self.Col):
                for i in range(1,abs(Row2-self.Row)):
                    if (Board[self.Row-i][self.Col+i] != "  "):
                        return False
            #if the move is down left
            else:
                for i in range(1,abs(Row2-self.Row)):
                    if (Board[self.Row-i][self.Col-i] != "  "):
                        return False
        #accessing the destination object
        obj = Board[Row2][Col2]

        # if empty destination
        if obj == "  ":
            return True
        # if the color is same, then return false
        if obj.Color == self.Color:
            return False
        return True
    

    # Move the Bishop object in that index and put "  " in previous index
    def Move(self,Row2,Col2,Board):
        # store the previus position in temp variables
        tempRow = self.Row
        tempCol = self.Col
        
        
        self.Moved += 1
        #changing the position of the piece
        self.Row = Row2
        self.Col = Col2
        
        
        # Move the Bishop object in that index and put "  " in previous index
        Board[Row2][Col2] = self
        Board[tempRow][tempCol] = "  "
        return Board
```

```python
#class of pawn piece
class Pawn(Piece):
    def __init__(self, name,row, col, importance, color, gui, moved):
        super().__init__(name,row, col, importance, color, gui, moved)


    # Validating pawn moves
    def Valid_Move(self,Row2,Col2,Board):


        #if move not forward, return false
        if (self.Color == "White" and Row2 > self.Row):
            # print("Cant move backwards for white")
            return False
        elif (self.Color == "Black" and Row2 < self.Row):
            # print("Cant move backwards for black")
            return False
        
        #if move is not 1 or 2 spaces forward, return false
        if abs(Row2-self.Row) > 2:
            # print("Cant move more than 2 spaces")
            return False
      

        #if white piece moves 2 piece forward and there is a piece in between, return false
        if (self.Color == "White" and abs(Row2-self.Row) == 2 and self.Row == 6 and Board[self.Row-1][self.Col] != "  "):    
            return False            


        #if black piece moves 2 piece forward and there is a piece in between, return false
        if (self.Color == "Black" and abs(Row2-self.Row) == 2 and self.Row == 1 and Board[self.Row+1][self.Col] != "  "):
            return False
        #if move is not 1 space forward, return false
        if (self.Color == "White" and abs(Row2-self.Row) > 1 and self.Row != 6):
            
            # print("Cant move more than 1 space for white")
            return False
        elif (self.Color == "Black" and abs(Row2-self.Row) > 1 and self.Row != 1):
            # print("Cant move more than 1 space for black")
            return False
        
        #if move is not 1 space sideways, return false
        if (abs(Col2-self.Col) > 1):
            # print("Cant move more than 1 space sideways")
            return False
        


        

        # if one move diagnol and no piece to capture, return false
        if (abs(Row2-self.Row) == 1 and abs(Col2-self.Col) == 1 and Board[Row2][Col2] == "  "):
            # print("No piece to capture")
            return False


        # if move forward and there is a piece to capture at diagnol of opposite color, return true
        if (abs(Row2-self.Row) == 1 and abs(Col2-self.Col) == 1 and Board[Row2][Col2] != "  " and Board[Row2][Col2].Color != self.Color):
            return True
        
        
        # if pawn moves forward and there is a piece in front, return false
        if (abs(Row2-self.Row) == 1 and abs(Col2-self.Col) == 0 and Board[Row2][Col2] != "  "):
            # print("There is a piece in front")
            return False
        
        # # if one move diagnol and the piece is of same color, return false
        # if (abs(Row2-self.Row) == 1 and abs(Col2-self.Col) == 1 and Board[Row2][Col2].Color == self.Color):
        #     print("Cant capture own piece")
        #     return False


        obj = Board[Row2][Col2]
        #if the destination is empty
        if obj == "  " and self.Col == Col2:
            return True
        #if the destination is not empty
        elif obj != "  " and obj.Color == self.Color:
            # print("Cant capture own piece")
            return False
        

        

    # Move the Pawn object in that index and put "  " in previous index
    def Move(self,Row2,Col2,Board):
        # store the previus position in temp variables
        tempRow = self.Row
        tempCol = self.Col
        self.Moved +=1


        self.Moved += 1
        #changing the position of the piece
        self.Row = Row2
        self.Col = Col2
        

        # Move the Pawn object in that index and put "  " in previous index
        Board[Row2][Col2] = self
        Board[tempRow][tempCol] = "  "
        return Board

        #if pawn reaches the end of the board, promote it
        if (self.Color == "White" and self.Row == 0):
            #ask what it wants to become
            print("Pawn has reached the end of the board, what would you like it to become (Queen, Rook, Bishop, Knight): ")
            choice = input()
            if choice == "Queen":
                Board[Row2][Col2] = Queen("Queen",Row2,Col2,9,"White","",0)   
            elif choice == "Rook":  
                Board[Row2][Col2] = Rook("Rook",Row2,Col2,5,"White","",0)
            elif choice == "Bishop":
                Board[Row2][Col2] = Bishop("Bishop",Row2,Col2,3,"White","",0)
            elif choice == "Knight":
                Board[Row2][Col2] = Knight("Knight",Row2,Col2,3,"White","",0)

        elif (self.Color == "Black" and self.Row == 7):
            #ask what it wants to become
            print("Pawn has reached the end of the board, what would you like it to become (Queen, Rook, Bishop, Knight): ")
            choice = input()
            if choice == "Queen":
                Board[Row2][Col2] = Queen("Queen",Row2,Col2,9,"Black","",0)   
            elif choice == "Rook":  
                Board[Row2][Col2] = Rook("Rook",Row2,Col2,5,"Black","",0)
            elif choice == "Bishop":
                Board[Row2][Col2] = Bishop("Bishop",Row2,Col2,3,"Black","",0)
            elif choice == "Knight":
                Board[Row2][Col2] = Knight("Knight",Row2,Col2,3,"Black","",0)
```

```python
# a function that takes in user input and converts it into Board indexes
def Get_Row_Col(r1,c1,r2,c2):
    DictRow = {
        "1":0,
        "2":1,
        "3":2,
        "4":3,
        "5":4,
        "6":5,
        "7":6,
        "8":7

    }
    DictCol = {
        "a":0,
        "b":1,
        "c":2,
        "d":3,
        "e":4,
        "f":5,
        "g":6,
        "h":7
    }
    r1 = DictRow[r1]
    c1 = DictCol[c1]
    r2 = DictRow[r2]
    c2 = DictCol[c2]
    return r1,c1,r2,c2
    
# Validating user input
def Validate_Index(r1,c1,r2,c2):
    RowList = ["1","2","3","4","5","6","7","8"]
    ColList = ["a","b","c","d","e","f","g","h"]
    if r1 in RowList and c1 in ColList and r2 in RowList and c2 in ColList:
        return True
    else:
        return False


```
